```python
import folium
import geopandas as gpd
import pandas as pd
from kakaoMap import KakaoMap
```

```python
df = pd.read_csv('./data/seoul_hagwon.csv', encoding='cp949')
df
```

```python
drop_condition  = df["분야명"] == "입시.검정 및 보습"
df_univ = df[drop_condition]

df_univ
```

```python
kakao = KakaoMap()
gangnam = kakao.convert_address_to_coords("서울특별시 중랑구 동일로151길 30")
gangnam
```

```python
kakao = KakaoMap()
for index, row in df_univ.iterrows():
  try:
    df_univ.loc[index, "위도"], df_univ.loc[index, "경도"] = kakao.convert_address_to_coords(row['도로명주소'])
  except:
    pass
```

```python
# file 내보내기
file_path = './data/hagwon_coords.csv'
df_univ.to_csv(file_path,index=False)
```

```python
# GeoJSON 파일 경로 설정
geojson_path = './HangJeongDong.geojson'  # GeoJSON 파일 경로를 지정하세요

# GeoJSON 데이터 로드
seoul_geojson = gpd.read_file(geojson_path)

# 중심 좌표 설정 (예: 강남구)
center = [37.5172, 127.0473]

# 지도 생성
m = folium.Map(location=center, zoom_start=11, tiles=None)

# 행정동 경계를 그리는 GeoJsonLayer 추가
# folium.GeoJson(seoul_geojson, name='seoul_dongs')

folium.Choropleth(
    geo_data=seoul_geojson,
    fill_color='#000',  # 색상 맵 (Yellow-Green)
    fill_opacity=0.2,
    line_opacity=0.7,
).add_to(m)

# 지도에 타일 레이어 추가 (OpenStreetMap)
folium.TileLayer('Stamen Toner').add_to(m)
```

```python
# 여러 개의 점을 지도에 추가 (아이콘 설정 적용)
for _, row in df_univ.iterrows():
    try:
       folium.Circle(location = [row['위도'], row['경도']],
              radius = 2,
              popup=row['학원명'],
              color='red'
       ).add_to(m)
    except:
       pass
```

```python
# LayerControl 추가
folium.LayerControl().add_to(m)

# HTML 파일로 저장
m.save('hagwon_map.html')
```
