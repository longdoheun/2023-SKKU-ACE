### 강남구 취소일 배제, 22-23 추출

```python
df = pd.read_csv('./data/gangnam.csv', encoding='cp949')
```

```python
df
```

```python
# 드랍할 조건 설정
drop_condition = df['계약일'] > 20211231

# 조건에 맞지 않는 행 드랍
df_22_23 = df[drop_condition]

df_22_23
```

```python
df_22_23 = df_22_23[df_22_23['취소일'].isna()]
df_22_23
```

```python
# file 내보내기
file_path = './data/gangnam_22to23.csv'
df_22_23.to_csv(file_path,index=False)
```

### 양천구 취소일 배제, 21 출력

```python
yangcheon = pd.read_csv('./data/yangcheon.csv', encoding='cp949')
yangcheon
```

```python
# 드랍할 조건 설정
drop_condition1 = (yangcheon['계약일'] > 20191231)
drop_condition2 = (yangcheon['계약일'] <= 20201231)
# 조건에 맞지 않는 행 드랍
yangcheon_21 = yangcheon[drop_condition1][drop_condition2]

yangcheon_21
```

```python
yangcheon_21 = yangcheon_21[yangcheon_21['취소일'].isna()]
yangcheon_21
```

```python
# file 내보내기
file_path = './data/yangcheon_20.csv'
yangcheon_21.to_csv(file_path,index=False)
```
